```python
from scipy.spatial.transform import Rotation
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
r = [Rotation.from_euler("ZYX", [yaw, 0, 0], degrees=True) for yaw in range(0, 360)]
```

```python
q = [rot.as_quat() for rot in r]
```

```python
plt.plot([quat[0] for quat in q], label='q0')
plt.plot([quat[1] for quat in q], label='q1')
plt.plot([quat[2] for quat in q], label='q2')
plt.plot([quat[3] for quat in q], label='q3')
plt.legend()
```

```python
eulers = [rot.as_euler('ZYX', degrees=True) for rot in r]
```

```python
plt.plot([rpy[0] for rpy in eulers], label='yaw')
plt.plot([rpy[1] for rpy in eulers], label='pitch')
plt.plot([rpy[2] for rpy in eulers], label='roll')
plt.legend()
```

```python
rotvecs = [rot.apply(np.array([1, 0, 0])) for rot in r]
plt.plot([vec[0] for vec in rotvecs], label='n/x')
plt.plot([vec[1] for vec in rotvecs], label='e/y')
plt.plot([vec[2] for vec in rotvecs], label='d/z')
plt.legend()
```
